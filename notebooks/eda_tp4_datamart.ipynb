## â±ï¸ Analyse temporelle

```python
engine = create_engine("postgresql://postgres:admin@localhost:15435/postgres")

df = pd.read_sql("SELECT * FROM fact_trips LIMIT 20", engine)

dim_datetime = pd.read_sql(
    "SELECT datetime_id, datetime FROM dim_datetime",
    engine
)


df['pickup_datetime_id'] = df['pickup_datetime_id'].astype(int)
dim_datetime['datetime_id'] = dim_datetime['datetime_id'].astype(int)


dim_datetime = dim_datetime.rename(columns={'datetime_id': 'pickup_datetime_id'})


df = pd.merge(
    df,
    dim_datetime,
    left_on='pickup_datetime_id',
    right_on='pickup_datetime_id',
    how='left'
)

df['datetime'] = pd.to_datetime(df['datetime'])
df['month']     = df['datetime'].dt.to_period('M')

import matplotlib.pyplot as plt
df.groupby('month').size().plot(kind='bar')
plt.title("Nombre de trajets par mois")
plt.ylabel("Nombre de trajets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## ðŸ”„ CorrÃ©lation entre variables

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("CorrÃ©lation entre variables numÃ©riques")
plt.show()
```
